# Split the Column TopScorer to TopScorer and Goals

```python
df['Goals'] = df['TopScorer'].str.split('-').str[1].str.strip().astype(int)       #split Goals from TopScorer
df['TopScorer'] = df['TopScorer'].str.split('-').str[0].str.strip()               #remove Goals from TopScorer
df.head()
```

## Reordering the columns

```python
cols = list(df.columns)                                         # Get the current column order
topscorer_index = cols.index('TopScorer')                       # Find the index of 'TopScorer'
cols.insert(topscorer_index + 1, cols.pop(cols.index('Goals'))) # Insert 'Goals' after 'TopScorer'
df = df[cols]                                                   # Reorder the DataFrame
df.head()
```

# TopScorer Goals by Year

```python
top_scorer_goals_by_year = df.groupby('Year')['Goals'].max()
plt.figure(figsize=(10, 6))
sns.lineplot(x=top_scorer_goals_by_year.index, y=top_scorer_goals_by_year.values, marker='o')
plt.title('Top Scorer Goals Trend Across Years')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.grid(True)
plt.show()
```
